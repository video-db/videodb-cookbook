## Step 4: AI-Powered Lyric Segmentation & Image Prompt Generation

This is where the magic happens! We use an LLM to:

### Lyric Processing
- **Fix ASR errors** - Correct obvious transcription mistakes
- **Segment into stanzas** - Group lyrics into natural phrases
- **Assign precise timestamps** - Each line gets start/end times
- **Optimize for display** - 4-10 words per line for readability

### Visual Generation
- **Create image prompts** - One detailed prompt per stanza
- **Match your aesthetic** - Honors your style description
- **Ensure readability** - Central space kept clear for text overlay
- **Maintain consistency** - Similar color palette and style throughout
- **Select font colors** - High-contrast colors for maximum readability

The AI analyzes the mood and theme of each stanza to generate visually appropriate backgrounds.

```python
import json

prompt = f"""
You are preparing lyric-video stanzas from an imperfect speech-to-text transcript.

Inputs:
1) Full transcript text (may contain errors / missing punctuation):
{transcript_text}

2) Word-level timed transcript (list of dicts with start/end/text/speaker; may include "-" silence markers):
{json.dumps(transcript_timed, ensure_ascii=False)}

3) Video Name: {video.name}

4) Video Length: {video.length}

4) User Style Request: {user_request}

Task:
- Produce a JSON array of lyric stanzas.
- Fix obvious ASR glitches lightly (missing punctuation, wrong word forms) ONLY when strongly implied by context.
  Do NOT invent new lyrics beyond what is clearly present.
- Merge words into natural lines/sentences suitable for lyric videos (not word-by-word).
- Split long lines; target ~4â€“10 words per line.
- Use timing: start = earliest word start in the stanza, end = latest word end in the stanza.
- Keep speaker labels. If a stanza mixes speakers, set speaker="Mixed".
- Make sure that the entire video length is considered.
- No time duration from 0 -> complete duration i.e. {video.length} is skipped
- If there is any section with no lyrics, return the text as [Music...] but give me a prompt. DO NOT LEAVE ANY IN-BETWEEN SEGMENTS EMPTY. YOU MUST GIVE A IMAGE GEN PROMPT FOR IT
- For the end of the video if there are no lyrics, don't fill it with [Music...] but with the name of the song and singer name instead.
- Please fill the empty spaces within
- Add:
  - color: choose a hex color per speaker (consistent across output)
  - font: choose a good lyric-video font name (e.g., Montserrat/Poppins/Inter)
- In the beginning of the lyrics: If there is time for the first piece of lyrics to come, display the name of the music video and the singer if available in the video name in the first few seconds as part of the lyrics
  - Example text: 'Song name\\nSinger'
  - If there is no time, display it with the lyrics itself: 'Song name\\nSinger/s\\n\\nlyrics1\\nlyric 2'

Image Generation Requirements:
- For EACH stanza, create an "image_prompt" field with a detailed text-to-image generation prompt.
- Each image_prompt must be FULLY SELF-CONTAINED and INDEPENDENT (no references to "previous image" or "same as before").
- The prompts must honor the user's style request: {user_request}
- Images should reflect the mood, theme, and emotional tone of the current stanza.
- CRITICAL: Prompts must NEVER include text, lyrics, names, words, letters, or any written language in the image.
- Images should be aesthetically beautiful but NOT overly distracting - they serve as backgrounds for text overlay.
- Maintain VISUAL CONSISTENCY across all prompts through:
  * Consistent color palette/scheme
  * Consistent artistic style
  * Consistent mood/atmosphere
  * Similar composition approach
- Composition guidelines:
  * Keep the center area relatively clear or with subtle gradients for text readability
  * Avoid overly busy or cluttered compositions
  * Use balanced, harmonious layouts
  * Consider leaving breathing room for text overlay
- Each prompt should specify:
  * The scene/subject matter relevant to the stanza
  * The emotional mood/atmosphere
  * Lighting and color tone
  * Artistic style (matching user_request)
  * Composition notes (e.g., "soft focus background", "centered negative space")
- Prompts should create a natural visual flow/story progression when images are shown in sequence
- Target aspect ratio: 16:9 (suitable for video)

Font Color Selection:
- For EACH stanza, select a "font_color" that ensures optimal readability against the generated background image.
- Choose from ONLY these options (provide hex code):
  * White: #FFFFFF
  * Black: #000000
- Consider the image's brightness, color palette, and overall tone when selecting.
- Light/bright backgrounds â†’ use Black (#000000), Dark Blue (#00008B), or Dark Green (#006400)
- Dark/dim backgrounds â†’ use White (#FFFFFF) or Yellow (#FFFF00)
- Ensure HIGH CONTRAST between font and background for maximum readability.
- Maintain consistency: if consecutive stanzas have similar image styles, try to keep the same font color unless contrast requires a change.

Task:
- Produce a JSON array of lyric stanzas.
- Each stanza must contain a "lines" array.
- A "line" is usually a natural phrase (4-8 words).
- Give each line its own "start" and "end" timestamps based on the word-level transcript.

Output format: JSON array ONLY, exactly:
[
  {{
    "stanza_start": float,
    "stanza_end": float,
    "image_prompt": str,
    "font_color": str,
    "lines": [
      {{
        "text": str,
        "start": float,
        "end": float
      }}
    ]
  }},
  ...
]
"""

stanzas = coll.generate_text(
    prompt=prompt,
    model_name="pro",
    response_type="json",
)
print(stanzas)
```

### Helper Function: Calculate Text Positioning

This function calculates the vertical offset for each line of lyrics to ensure they're centered and evenly spaced on screen.

It handles multi-line stanzas gracefully, keeping everything readable and balanced.

```python
def calculate_y_offsets(num_lines, line_height=60, timeline_height=720):
    offsets = []
    start_pixel_offset = -((num_lines - 1) * line_height) / 2

    half_height = timeline_height / 2

    for i in range(num_lines):
        pixel_y = start_pixel_offset + (i * line_height)
        normalized_y = pixel_y / half_height
        offsets.append(normalized_y)
    return offsets
```

### Extract Stanza Data

Retrieve the generated stanzas from the LLM response for video composition.

```python
stanza_items = stanzas.get("output")
```

## Step 5: Generate Background Images

Now we generate the actual images! For each stanza, we create a custom 16:9 background image based on the AI-generated prompts.

**Key features:**
- Optimized for text overlay with central negative space
- Matches your aesthetic description
- Smooth visual transitions between stanzas
- High-quality images suitable for video

This cell uses parallel processing to generate multiple images simultaneously for faster execution.

```python
from concurrent.futures import ThreadPoolExecutor, as_completed

def _gen_one(i, stanza):
    img = coll.generate_image(prompt=stanza["image_prompt"], aspect_ratio="16:9")
    return i, img.id

max_workers = 10

results = {}
with ThreadPoolExecutor(max_workers=max_workers) as ex:
    futures = [ex.submit(_gen_one, i, stanza) for i, stanza in enumerate(stanza_items)]
    for fut in as_completed(futures):
        i, image_id = fut.result()
        results[i] = image_id

for i, stanza in enumerate(stanza_items):
    stanza["image_id"] = results[i]
```

## Step 6: Compose the Final Lyric Video

Time to bring everything together! We'll build a multi-layered timeline with:

### Layer 1: Audio Track
- Original music from the video
- Full volume for clear playback
- Hidden video (opacity=0) so only audio plays

### Layer 2: Background Images
- Generated images displayed in sequence
- Fade transitions between stanzas
- Perfect sync with lyric timing
- Music segment optimization for smooth flow

### Layer 3: Animated Lyrics
- Word-by-word granular line reveal
- Precisely timed to match the audio
- Custom font styling with borders for readability
- Fade-in animations for smooth appearance

The result is a professional lyric video with synchronized text, beautiful visuals, and the original audio track.

```python
from videodb.editor import (
    VideoAsset, ImageAsset, TextAsset, Font,
    Clip, Track, Timeline, Transition,
    Position, Fit, Offset, Alignment, HorizontalAlignment, VerticalAlignment, Border
)

total_duration = max(float(s["stanza_end"]) for s in stanza_items)

timeline = Timeline(connection)
timeline.resolution = "1280x720"
timeline.background = "#000000"

def optimize_stanza_transitions(stanzas):
    if not stanzas:
        return []
    optimized = []
    for i in range(len(stanzas)):
        current = stanzas[i]
        is_music = current.get("lines") and "[Music...]" in current["lines"][0].get("text", "")
        if is_music and i > 0:
            prev = optimized[-1]
            music_duration = current["stanza_end"] - current["stanza_start"]
            if music_duration < 2.0:
                prev["stanza_end"] = current["stanza_end"]
                if prev.get("lines"):
                    prev["lines"][-1]["end"] = current["stanza_end"]
                continue
            else:
                extension_time = 1.0
                prev["stanza_end"] += extension_time
                if prev.get("lines"):
                    prev["lines"][-1]["end"] += extension_time
                current["stanza_start"] += extension_time
                if current.get("lines"):
                    current["lines"][0]["start"] += extension_time
                if current["stanza_start"] >= current["stanza_end"]:
                    continue
                optimized.append(current)
        else:
            optimized.append(current)
    return optimized

stanza_items = optimize_stanza_transitions(stanza_items)

# 1) AUDIO track (original video, hidden but audible)
audio_track = Track(z_index=0)
audio_track.add_clip(
    start=0.0,
    clip=Clip(
        asset=VideoAsset(id=video.id, start=0, volume=1.0),
        duration=video.length,
        fit=Fit.crop,
        position=Position.center,
        opacity=0.0,
        scale=1.0,
    ),
)
timeline.add_track(audio_track)

images_track = Track(z_index=1)
FADE_SEC = 0.35

for s in stanza_items:
    start = float(s["stanza_start"])
    end = float(s["stanza_end"])
    duration = max(0.05, end - start)

    images_track.add_clip(
        start=start,
        clip=Clip(
            asset=ImageAsset(id=s["image_id"]),
            duration=duration,
            fit=Fit.crop,
            position=Position.center,
            transition=Transition(in_="fade", out="fade", duration=min(FADE_SEC, duration / 2)),
            opacity=1.0,
            scale=1.0,
        ),
    )

timeline.add_track(images_track)

# 3) LYRICS track (Granular Line Reveal)
lyrics_track = Track(z_index=2)
LINE_HEIGHT = 32
border = Border(color="#000000", width=1.5)

for s in stanza_items:
    lines = s.get("lines", [])
    y_offsets = calculate_y_offsets(len(lines), line_height=LINE_HEIGHT)
    stanza_end = float(s["stanza_end"])

    for idx, line in enumerate(lines):
        line_start = float(line["start"])
        line_duration = max(0.1, stanza_end - line_start)

        text_asset = TextAsset(
            text=line["text"],
            font=Font(
                family="Bebas Neue",
                size=36,
                color="#FFFFFF",
                weight=700
            ),
            border=border,
            alignment=Alignment(
                horizontal=HorizontalAlignment.center,
                vertical=VerticalAlignment.center
            )
        )

        lyrics_track.add_clip(
            start=line_start,
            clip=Clip(
                asset=text_asset,
                duration=line_duration,
                position=Position.center,
                offset=Offset(x=0, y=y_offsets[idx]),
                transition=Transition(in_="fade", duration=0.5),
                opacity=1.0
            )
        )

timeline.add_track(lyrics_track)

# 4) Render
stream_url = timeline.generate_stream()
print("Stream URL:", stream_url)
```
